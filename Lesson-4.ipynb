```r
pacman::p_load("tidyverse", "ggpubr")
```

```r
df <- read_table2("STAT501_Lesson01/bloodpress.txt")
head(df)
```

```r
slm.age <- lm(BP ~ Age, data=df)
plot(BP ~ Age, data=df)
abline(slm.age)
summary(slm.age)
```

```r
plot(slm.age$residuals ~ slm.age$fitted.values)
abline(h=0)
```

```r
plot(slm.age$residuals ~ df$Dur)
abline(h=0)
```

```r
plot(slm.age$residuals ~ df$Weight) ## adding more predictor variables is worthwhile.
abline(h=0)
```

```r
plot(slm.age)
```

```r
mlm <- lm(BP ~ Age + Weight + Stress, data=df)
summary(mlm)
```

```r
plot(mlm)
```

```r
trd.df <- read_table2("STAT501_Lesson01/treadmill.txt")
head(trd.df)
```

```r
mlm <- lm(vo2 ~ dur + age, data=trd.df)
summary(mlm)
```

```r
plot(vo2 ~ dur, data=trd.df)
```

```r
cor(trd.df$vo2, trd.df$dur) ^2
```

```r
plot(vo2 ~ age, data=trd.df)
cor(trd.df$vo2, trd.df$age) ^2
```

```r
plot(dur ~ age, data=trd.df)
cor(trd.df$dur, trd.df$age) ^2
```

```r
slm.dur <- lm(vo2 ~ dur, data=trd.df)
summary(slm.dur)
```

```r
plot(slm.dur$residuals ~ trd.df$age) ## adding age variable is not helpful. no patterns!
```

```r
slm.age <- lm(vo2 ~ age, data=trd.df)
summary(slm.age)
```

```r
plot(slm.age$residuals ~ trd.df$dur) ## adding dur variable would be helful. Found linear patterns!
```
